# Multi-Plots in Seaborn

## Fason Grid

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Visualization-Workshop/master/Datasets/salary.csv')[:1000]
g = sns.FacetGrid(data, col='District')
g.map(plt.scatter, 'Salary', 'Age')
```
