# Linear Regression for Animal Attribute Relations

```
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
mydata= pd.read_csv('https://raw.githubusercontent.com/PacktWorkshops/The-Data-Visualization-Workshop/master/Datasets/anage_data.csv')
```

```
longevity = 'Maximum longevity (yrs)'
mass = 'Body mass (g)'
data = mydata[mydata['Class'] == 'Mammalia']
data = data[np.isfinite(data[longevity]) & np.isfinite(data[mass]) & (data[mass] < 200000)]
```

```
# Create figure
sns.set()
plt.figure(figsize=(8, 6), dpi=200)
# Create a scatter plot
sns.regplot(mass, longevity, data=data)
# Show plot
plt.show()
```
